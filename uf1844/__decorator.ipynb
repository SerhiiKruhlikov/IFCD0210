```python
from typing import Callable

def empty_deco(func: Callable):
    def wrapper():
        res = func()
        return res
    return wrapper
```

```python
@empty_deco
def my_func():
    return 124
```

```python
print(my_func())
```

```python
from typing import Callable
import time


def timer_deco(func: Callable):
    def wrapper():
        start = time.time()
        res = func()
        end = time.time()
        print(f"исполнение заняло {end - start}")
        return res
    return wrapper


@timer_deco
def my_func():
    time.sleep(1.5)
    return 124

my_func()
```

```python
from typing import Callable
import time


def param_timer_deco(func: Callable):
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()
        print(f"исполнение заняло {end - start}")
        return res
    return wrapper


@param_timer_deco
def my_func(sleep_sec: int):
    time.sleep(sleep_sec)
    return 124

my_func(3)
```

```python
from typing import Callable
import time
import functools


def limit_calls(limit: int):
    def wrapper(func: Callable):
        @functools.wraps(func)
        def inner(*args, **kwargs):
            nonlocal limit
            if limit == 0:
                print("нельзя вызвать функцию")
                return
            res = func(*args, **kwargs)
            limit -= 1
            return res
        return inner
    return wrapper


@limit_calls(2)
def my_func(sleep_sec: int):
    """Очень важный докстинг не потерять его"""
    time.sleep(sleep_sec)
    return 124

print(my_func.__name__)
print(my_func.__doc__)
my_func(1)
```

```python
from typing import Coroutine
import asyncio


def deco(coroutine: Coroutine):
    async def wrapper(*args, **kwargs):
        res = await coroutine(*args, **kwargs)
        return res
    return wrapper
 

@deco
async def my_async_func():
    await asyncio.sleep(.5)
    return 1

await my_async_func()
```

```python
from typing import Callable
import functools

def deco(func: Callable):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"func {func.__name__} args {args} kwargs {kwargs}")
        result = func(*args, **kwargs)
        return result
    return wrapper


@deco
def my_foo(*args, **kwargs):
    return args, kwargs


my_foo(2, 3, name="Serhii", age=48)
```
