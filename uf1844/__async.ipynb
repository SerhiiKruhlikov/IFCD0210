```python
import time
import requests


def fetch_url(url):
    print(f"I'm starting to download {url}")
    response = requests.get(url)
    print(f"Finished downloading {url}")
    return response.text


start = time.time()
fetch_url("https://apple.es")
fetch_url("https://apple.com")
print(f"Time: {time.time() - start}")    
```

```python
import asyncio
import aiohttp
import time
import ssl


async def fetch_url(session, url):
    print(f"I'm starting to download {url}")
    async with session.get(url) as response:
        text = await response.text()
        print(f"Finished downloading {url}")
        return text

async def main():
    ssl_context = ssl.create_default_context()
    ssl_context.check_hostname = False
    ssl_context.verify_mode = ssl.CERT_NONE
    
    async with aiohttp.ClientSession(connector=aiohttp.TCPConnector(ssl=ssl_context)) as session:
        task1 = asyncio.create_task(fetch_url(session, "https://apple.es"))
        task2 = asyncio.create_task(fetch_url(session, "https://apple.com"))
        
        result1 = await task1
        result2 = await task2
        
        print(f"Длина первого ответа: {len(result1)} символов")
        print(f"Длина второго ответа: {len(result2)} символов")


start = time.time()
await main()
print(f"Lead time: {time.time() - start} seconds")  
```
