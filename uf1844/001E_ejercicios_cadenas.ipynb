1. Escriba un programa que solicite al usuario que ingrese una cadena. El programa debería mostrar lo siguiente:
- (a) El número total de caracteres de la cadena
- (b) La cadena repetida 10 veces
- (c) El primer carácter de la cadena (recuerde que los índices de cadena empiezan en 0)
- (d) Los tres primeros caracteres de la cadena
- (e) Los tres últimos caracteres de la cadena
- (f) La cadena al revés
- (g) El séptimo carácter de la cadena si la cadena es lo suficientemente larga; en caso contrario, un mensaje
- (h) La cadena sin el primer y el último carácter
- (i) La cadena en mayúsculas
- (j) La cadena con cada a reemplazada por una e
- (k) La cadena con cada letra reemplazada por un espacio

```python
cadena = input('ingrese una cadena:')
print('(a) El número total de caracteres de la cadena:', len(cadena))
print('(b) La cadena repetida 10 veces:', cadena * 10)
print('(c) El primer carácter de la cadena (recuerde que los índices de cadena empiezan en 0):', cadena[0])
print('(d) Los tres primeros caracteres de la cadena:', cadena[:3])
print('(e) Los tres últimos caracteres de la cadena:', cadena[-3:])
print('(f) La cadena al revés:', cadena[::-1])
g = '(g) El séptimo carácter de la cadena si la cadena es lo suficientemente larga; en caso contrario, un mensaje'
if len(cadena) <= 7:
    print(f'{g}: La cadena no tiene suficiente carácteres!')
else:
    print(f'{g}:', cadena[6])
print('(h) La cadena sin el primer y el último carácter:', cadena[1:-1:])
print('(i) La cadena en mayúsculas:', cadena.upper())
print('(j) La cadena con cada a reemplazada por una e:', cadena.replace('a', 'e'))
print('(k) La cadena con cada letra reemplazada por un espacio:', ' '.join(cadena))
print('(k) La cadena con cada letra reemplazada por un espacio:', '*' * len(cadena))
```

2. Una forma sencilla de estimar el número de palabras en una cadena es contar el número de espacios en la cadena. <br>Escriba un programa que solicite al usuario una cadena y devuelva una estimación de cuántas palabras tiene.

```python
frase = 'Lorem   ipsum   dolor   sit amet, ..,,    consectetur adipisicing elit ,,, '
excluyendo = (',','.',':',';','?','!',)

for caracter in excluyendo:
    frase = frase.replace(caracter, '')

lista = []

for palabra in frase.split(' '):
    if len(palabra) > 0:
        lista.append(palabra)

print('Variant 1:', len(lista), lista)

print('Variant 2:', len([palabra for palabra in frase.split(' ') if len(palabra) > 0]), lista)
```
