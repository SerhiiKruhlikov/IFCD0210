```python
def mostrar_config(**kwargs):
    lista = list(kwargs.items())
    
    print(lista)
    
    USUARIO = kwargs.get('usr', 'anonim')
    IP = kwargs.get('ip', '127.0.0.1')
    
    print(USUARIO, IP)
    
    for k, v in list(kwargs.items()):
        print(f"{k}: {v}")

mostrar_config(usr='teo', ip='1.2.3.4', puerto='8080', altura=1.77)

d = {
    'usr': 'teo',
    'ip': '1.2.3.4',
    'puerto': '8080',
    'altura': 1.77,
}

mostrar_config(**d)
```

```python
def log(mensaje, *args, **kwargs):
    nivel = kwargs.get('nivel', 'DEBUG')
    prefijo = kwargs.get('prefijo', '')
    
    msj_formato = f"[{nivel}]: {prefijo} - {mensaje}"
    
    return msj_formato

log("Error de configuracion", nivel='DEBUG')
```

```python
def http_get(url, **kwargs):
    timeout = kwargs.get('timeout', 5)
    headers = kwargs.get('headers', {})
    params = kwargs.get('params', {})
    
    return {
        'url': url,
        'timeout': timeout,
        'headers': headers,
        'params': params,
    }


http_get('http://google.com', timeout=10, params={'page': 2}, headers={})
```
