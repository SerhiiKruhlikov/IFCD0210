```python
peso = float(input("Peso: "))
altura = float(input("Altura: "))
imc = peso / altura ** 2

if imc < 18:
    print("IMC es: ", imc, ", es Bajo")
elif 18 < imc < 25:
    print("IMC es: ", imc, ", es Normal")
else:
    print("Sobrepeso!")
```
