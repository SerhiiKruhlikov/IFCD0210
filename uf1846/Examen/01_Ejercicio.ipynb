#### Dado el siguinte sistems:
* Un navegador muestra una web.
* Un servidor procesa peticiones.
* Una base de datos almacen usuarios.

#### Responde:
1. ¿Qué parte es el frontend?
2. ¿Qué parte es el backend?
3. ¿Es una arquitectura distribuida? ¿Por qué?

1. El Frontend es el navegador, ya que se encarga de mostrar la interfaz y de la interacción con el usuario.
2. El Backend es el servidor (que procesa las peticiones) y la base de datos (que almacena la información), ya que ejecutan la lógica de procesamiento y el almacenamiento de datos en el lado del servidor.
3. Sí, es una arquitectura distribuida, porque los componentes del sistema (navegador, servidor, base de datos) funcionan en diferentes nodos, interactuando entre sí a través de la red. El navegador se encuentra en la máquina del cliente, mientras que el servidor y la base de datos están en el lado del servidor, lo que cumple con los principios de los sistemas distribuidos.

#### Tienes esta API:
GET /usuarios
POST /usuarios
GET /usuarios/5
DELETE /usuarios/5
    
#### Responde:
1. ¿Qué hace cada endpoint?
2. ¿Cuál usarías para crear un usuario?
3. ¿Cuál usarías para eliminarlo?

1. Qué hace cada endpoint:
    GET /usuarios — obtiene la lista de todos los usuarios.
    POST /usuarios — crea un nuevo usuario.
    GET /usuarios/5 — obtiene los datos del usuario con el identificador 5.
    DELETE /usuarios/5 — elimina al usuario con el identificador 5.
2. Para crear un usuario se utiliza POST /usuarios.
3. Para eliminar un usuario se utiliza DELETE /usuarios/5 (donde 5 es el identificador del usuario que se desea eliminar).

#### Diseña una API para una aplicación de tareas (to-do list).
Define al menos 4 endpoints:
* Obtener tareas
* Crear tarea
* Actualizar tarea
* Eliminar tarea

API para una aplicación de lista de tareas (to-do list):
1. Obtener tareas — GET /tareas
   Retorna la lista de todas las tareas.
   Adicionalmente se puede agregar:
   GET /tareas/{id} — Obtener una sola tarea por su ID.
   GET /tareas?complete=true — Filtrado por estado (completada).
2. Crear una tarea — POST /tareas
   Recibe un JSON con los datos de la tarea, por ejemplo:
   ```{"title": "Titulo la tarea", "complete": false}```.
   Crea una nueva tarea y la devuelve con un ID asignado.
3. Actualizar una tarea — PUT /tareas/{id} o PATCH /tareas/{id}
   Recibe un JSON con los campos a actualizar, por ejemplo:
   ```{"title": "Nuevo título la tarea", "complete": true}```.
   Actualiza la tarea con el identificador especificado.
4. Eliminar una tarea — DELETE /tareas/{id}
   Elimina la tarea con el identificador especificado.
