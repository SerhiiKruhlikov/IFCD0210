#### Dado el siguinte sistems:
* Un navegador muestra una web.
* Un servidor procesa peticiones.
* Una base de datos almacen usuarios.

#### Responde:
1. ¿Qué parte es el frontend?
2. ¿Qué parte es el backend?
3. ¿Es una arquitectura distribuida? ¿Por qué?

1. Фронтэнд (браузер) отвечает за отображение интерфейса и взаимодействие с пользователем посредством браузера.
2. Бэкэнд (сервер) отвечает за логику обработки запросов и хранение данных на стороне сервера.
3. Да, это распределенная архитектура, так как компоненты системы работают на разных узлах, взаимодействуя между собой через сеть. Браузер находится на клиентской машине, а сервер и база данных — на серверной стороне, что соответствует принципу распределенной системы.
